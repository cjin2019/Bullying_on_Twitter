```
import urllib2 

indices = []

for i in range(0, df_data.shape[0]):
  if i % 500 == 0:
    print(i)
  try:
    if urllib2.urlopen("https://twitter.com/statuses/" +  str(df_data['Tweet ID'][i])).code:
      continue
  except urllib2.HTTPError, e:
    indices.append(i)
```
