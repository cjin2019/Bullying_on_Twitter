```
### creating list of the data

tweets = []
original_tweets = []
tweets_id = []

for i in range(1):
  tweets.append(l[i]['text'].encode('ascii', 'ignore'))
  tweets_id.append(l[i]['id'])
  original_tweets.append(tweets[i])
  
```

```
### get rid of the punctuation 

import string
punctuation = string.punctuation

### replaces the #, @, ",", "."
for i in range(len(tweets)):
  curr_str = tweets[i]
  
  for punc in punctuation:
    if punc == ":" or punc == ";" or punc == ")" or punc == "(" or punc == "|":
      continue
    curr_str = curr_str.replace(punc, "")
    
  tweets[i] = curr_str
```

```
### lowercase 
for i in range(len(tweets)):
  tweets[i] = tweets[i].lower()
```

```
tweets
```

# TRANSFORM INTO VECTORS

```
#### see for each tweet the word is in keyed vector
for i in range(len(tweets)):
  word_split = tweets[i].split()
  in_keyed_vector = []
  for w in word_split:
    try:
      arr = kv[w]
      in_keyed_vector.append(w)
    except Exception:
      print('Exception')
      continue
  new_word = ' '.join(in_keyed_vector)
  tweets[i] = new_word
  #print(new_word)
  
  
```

```
tweets
```

```
#### make to_Vec
maxL = 31
### creating the array for the vectors
to_Vec = np.zeros((len(tweets), maxL, 100))

for i in range(len(tweets)):
  s = tweets[i]
  words = s.split()
  
  npArr = np.zeros((maxL, 100)) ## shape of the numpy 
  for j in range(len(words)):
    npArr[j] = kv[words[j]]
  to_Vec[i] = npArr
```

```
to_Vec
```

# PREDICTING

```
probs = clf.predict(to_Vec)
bullying_trace = [np.argmax(y) for y in clf.predict(to_Vec)]
probs, bullying_trace
```
